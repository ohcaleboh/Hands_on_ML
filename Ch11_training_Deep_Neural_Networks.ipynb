# Vanishing/Exploding Gradient Problems

Deep learning networks suffer from unstable gradients --as they get to lower layers, they can stagnate, failing to converge, hence *vanishing gradient* OR the gradient can grow bigger until the weight updates are so large that the algo diverges *exploding gradeints*. 

note: this observation is actually why deep learning nets were put on haitus for some time. 



Look at the log activation function above. when inputs become large, the function approaches 0 or 1. w/ deratives v close to 0. so, when backpropagation kicks in, there is pretty much no gradeint as it keeps getting diluted as backprop progresses down from the top layers. 

# Glorot and He initialization 


The solution propsoed by Glorot and Bengio is to try to make the variance of the outputs of each layer equal to the input's variance. ditto for graidents before and after flowing thru a layer in the reverse direction.

It is not actually possible to do both perfectly, so a compromise made by initializing the connection weights of each layer randomly. 



To use this implementation, you just set the init parameters accordingly (see the book for specifics). Overall, the reason we do this is so that neural nets can be sped up by making sure that the input and output signal for forward and back prop for each of the neurons is not too large or too small --exploding/vanishing gradients respectively.


# Non saturating Activation Functions

It might also be helpful to use different activation functions --other than log regression-- at each layer in order to train better NNs. The reason being that sometimes with the default ReLU activation function, you risk many neurosn 'dying' which just means the vanishing problem --the output is zero. 

Instead you can use a different activation fucntion that circumvents this problem --the ELU function below. Or you can use a 'leaky ReLU activation function' which is just a mod to the hyperparaemter. 



the main drawback is that ELU function is slower. 

So, how do you chose which activation fucntsion to use? 
This is a prioritization: > ELU > leaky ReLU (and its variants) > ReLU > tanh > logistic. 

# Batch Normalization 

Batch normalization is a method of standardizing the input and output at each layer. It removes the likelihood of exploding/vanishing gradients. But the tradeoff is that it takes longer to train the model (i.e. it is more complex).

You can find examples using the keras package. 

# Gradient Clipping

Another way to lessen the exploding gradientn problem is to use gradient clipping. This method is usually used in recurrent neural networks --typically you should just use BN.
